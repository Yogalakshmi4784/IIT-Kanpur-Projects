# Hypothesis Testing

### Older individuals may not possess the same level of technological proficiency and may, therefore, lean toward traditional in-store shopping preferences.

```python
bins = [0, 30, 45, 60, 100]
labels = ['18-29', '30-44', '45-59', '60+']

# Bin the ages
df['Age_Group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

```python
print('kurt',df['NumStorePurchases'].kurt())
print('skew',df['NumStorePurchases'].skew())

print('kurt',df['NumWebPurchases'].kurt())
print('skew',df['NumWebPurchases'].skew())
```

```python
# Lets go for parametric tests- Anova since we have more than 2 categories of age groups for comparison
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df_melted = pd.melt(df, id_vars=['Age_Group'], value_vars=['NumWebPurchases', 'NumStorePurchases'],
                    var_name='Purchase_Type', value_name='Count')

# Perform ANOVA
model = ols('Count ~ C(Age_Group) * C(Purchase_Type)', data=df_melted).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
```

```python
anova_table
```

# Observations

- The p-value for C(Age_Group) is 2.641667e-17. Since p-value is lesser than .05 there are significant differences in the Web store purchase and online store purchase across different age groups. 

- Hence Null hypothesis can be rejected saying that Older individuals may not possess the same level of technological proficiency and may, therefore, lean toward traditional in-store shopping preferences.

### Customers with children likely experience time constraints, making online shopping a more convenient option.

#### Null Hypothesis - Customers with children does not make any difference with other customers without children

#### Alternate Hypothesis- Customers with children likely experience time constraints, making online shopping a more convenient

```python
df['NumWebPurchases'].plot(kind='density')
```

```python
df['NumWebPurchases'].kurt()
```

```python
df['NumWebPurchases'].skew()
```

```python
# Since skewness is lesser than 2 and categories is equal to 2. lets conduct parametric test- ttest
```

```python
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
no=df[df['children']==0]
yes=df[df['children']>0]
ttest_ind(no.NumWebPurchases,yes.NumWebPurchases,equal_var=False)
```

## Observations

- Since p value < .05, Null hypothesis can be rejected saying Customers with children likely experience time constraints, making online shopping a more convenient way

### Does the United States significantly outperform the rest of the world in total purchase volumes?

```python
df['Tot_Purchases'].plot(kind='density')
```

```python
# Since this is a non parametric test and the categories are 2, we use Mann whitney
# Null Hypothesis - All the total purchase volumes are equal 
# Alternate Hypothesis - Total purchase volumes of US outperform the rest of the world
```

```python
US_df=df[df['Country']=='US']['Tot_Purchases']
Other_df=df[df['Country']!='US']['Tot_Purchases']
```

```python
df['Country'].value_counts()
```

```python
from scipy.stats import mannwhitneyu
stat, p_value = mannwhitneyu(US_df, Other_df)
```

```python
stat, p_value
```

# Observations

- Since p value is lesser than .05, reject the null hypothesis
- Total purchase volumes of US outperform the rest of the world
